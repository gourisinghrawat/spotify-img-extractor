```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import requests
import os
import re  # Added to remove special characters

# Spotify API credentials (replace with your own)
CLIENT_ID = "5fb41b62ddd1436d9417f7f490a1f2c6"
CLIENT_SECRET = "79de6c781f3f44b886c1d392fb1806ca"

# Initialize Spotify API client
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET))

def sanitize_filename(filename):
    """Removes invalid characters from filenames."""
    return re.sub(r'[<>:"/\\|?*]', '', filename)  # Removes special characters

def download_image(url, filename, folder="covers"):
    """Downloads the album cover image."""
    os.makedirs(folder, exist_ok=True)
    filename = sanitize_filename(filename)  # Sanitize filename
    filepath = os.path.join(folder, filename)

    response = requests.get(url)
    if response.status_code == 200:
        with open(filepath, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded: {filename}")
    else:
        print(f"Failed to download {filename}")

def get_playlist_tracks(playlist_url):
    """Fetches tracks from a Spotify playlist and downloads their album covers."""
    playlist_id = playlist_url.split("/")[-1].split("?")[0]
    results = sp.playlist_tracks(playlist_id)
    tracks = results['items']

    for track in tracks:
        track_info = track['track']
        song_name = track_info['name']
        artist_name = track_info['artists'][0]['name']
        album_cover_url = track_info['album']['images'][0]['url']

        # Save as "SongName_ArtistName.jpg"
        filename = f"{song_name}_{artist_name}.jpg"
        download_image(album_cover_url, filename)

if __name__ == "__main__":
    playlist_url = input("Enter the Spotify playlist URL: ")
    get_playlist_tracks(playlist_url)
```

```python
from PIL import Image
import os

def stitch_images(folder_path, output_folder="combine"):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    
    # Get list of image files in the folder
    image_files = [f for f in os.listdir(folder_path) if f.endswith(('png', 'jpg', 'jpeg'))]
    
    if len(image_files) < 4:
        raise ValueError("Need at least 4 images to stitch")
    
    # Process images in batches of 4
    for i in range(0, len(image_files), 4):
        batch = image_files[i:i+4]
        
        # Ensure we have exactly 4 images in the batch
        if len(batch) < 4:
            break
        
        images = [Image.open(os.path.join(folder_path, img)) for img in batch]
        
        # Resize images to the same size (use the size of the first image as reference)
        width, height = images[0].size
        images = [img.resize((width, height)) for img in images]
        
        # Create a blank canvas for stitching (2x2 grid)
        stitched_width = width * 2
        stitched_height = height * 2
        stitched_image = Image.new('RGB', (stitched_width, stitched_height))
        
        # Paste images onto the canvas
        stitched_image.paste(images[0], (0, 0))
        stitched_image.paste(images[1], (width, 0))
        stitched_image.paste(images[2], (0, height))
        stitched_image.paste(images[3], (width, height))
        
        # Save the final stitched image
        output_path = os.path.join(output_folder, f"stitched_{i//4 + 1}.jpg")
        stitched_image.save(output_path)
        print(f"Stitched image saved as {output_path}")

# Example usage
stitch_images("covers")
```
